Type casting in Python refers to converting one data type into another. This is useful when you need to perform operations that require specific data types or when you want to change the data type of a variable for compatibility purposes. Python supports both implicit and explicit type casting.
Implicit Type Casting
Implicit type casting is performed automatically by Python without any user intervention. It happens when Python converts a smaller data type into a larger data type to prevent data loss.
Explicit Type Casting
Explicit type casting, also known as type conversion, requires the programmer to use built-in functions to convert data types manually

```python
x = 5
y = 3.2


result = x + y

print(result)
print(type(result))
```

```python
num_str = "123"
num_int = int(num_str)

print(num_int)
print(type(num_int))
```

Python provides several types of loops that allow you to execute a block of code repeatedly, which is essential for tasks like iterating over sequences or performing repeated actions. The primary types of loops in Python are:

for loop
while loop
Nested loops
Each loop type has specific use cases and is suited for different scenarios. Let's explore them with examples.

1. for Loop
The for loop is used to iterate over a sequence (e.g., list, tuple, string, or range) and execute a block of code for each element.

Use Cases:

Iterating through items in a list or a dictionary.
Repeating actions a specific number of times using the range() function.
Processing elements in a sequence for tasks like data filtering or aggregation.

2. while Loop
The while loop is used to execute a block of code as long as a condition is True. It is particularly useful when you do not know in advance how many times you need to iterate.

Use Cases:

Repeating code until a certain condition is met (e.g., user input validation).
Implementing counters or time-based conditions.
Executing loops with conditions that might change dynamically during runtime.

Nested Loops
A nested loop is a loop inside another loop. Both for and while loops can be nested. Nested loops are useful for working with multi-dimensional data structures, like lists of lists or grids.

Use Cases:

Iterating over a matrix or 2D array.
Generating combinations or permutations.
Handling nested structures, such as nested dictionaries or complex data

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
for i in range(5):
    print(i)
```
